```python
# ==============================================================================
# 34 St-Hudson Yards Subway Fare Hikes & Ridership Elasticity Timeline (1915 - 2026)
# Featuring Historical Line Padding & Neighborhood Income Context
# ==============================================================================

# Install required libraries
!pip install -q pandas matplotlib seaborn requests plotly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json
from datetime import datetime

# Set up charting style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [14, 8]
plt.rcParams['figure.dpi'] = 100

print("✅ Libraries successfully imported and environment configured!")

## Step 1: Initialize Historical Station-Specific Ridership Data (Expanded back to 1915)

# Historical data matrix scaled specifically for the 34 St-Hudson Yards Station.
# Tracked in millions of annual passenger entries, with 0.0 entries prior to 2015
# to show the historical context of when the overall 7 line existed without this stop.
raw_hudson_yards_data = [
    {"year": 1915, "nominal_fare": 0.05, "annual_riders_millions": 0.0, "event": "7 Line Opens (Queensboro Plaza to LIC)"},
    {"year": 1928, "nominal_fare": 0.05, "annual_riders_millions": 0.0, "event": "7 Line Reaches Flushing-Main St"},
    {"year": 1939, "nominal_fare": 0.05, "annual_riders_millions": 0.0, "event": "1939 World's Fair Surge on 7 Line"},
    {"year": 1948, "nominal_fare": 0.10, "annual_riders_millions": 0.0, "event": "First Subway Fare Hike (+100%)"},
    {"year": 1953, "nominal_fare": 0.15, "annual_riders_millions": 0.0, "event": "Subway Tokens Introduced"},
    {"year": 1964, "nominal_fare": 0.15, "annual_riders_millions": 0.0, "event": "World's Fair Returns to Queens"},
    {"year": 1975, "nominal_fare": 0.50, "annual_riders_millions": 0.0, "event": "NYC Fiscal Crisis"},
    {"year": 1985, "nominal_fare": 0.90, "annual_riders_millions": 0.0, "event": "R62A Fleet Rollout on 7 Line"},
    {"year": 1995, "nominal_fare": 1.50, "annual_riders_millions": 0.0, "event": "MetroCard System-Wide Integration"},
    {"year": 2003, "nominal_fare": 2.00, "annual_riders_millions": 0.0, "event": "Tokens Discontinued"},
    {"year": 2015, "nominal_fare": 2.75, "annual_riders_millions": 1.4, "event": "34 St-Hudson Yards Station OPENS (Sept 13)"},
    {"year": 2016, "nominal_fare": 2.75, "annual_riders_millions": 2.7, "event": "First Full Year of Operations"},
    {"year": 2017, "nominal_fare": 2.75, "annual_riders_millions": 3.1, "event": "35th St Station Entrance Opens"},
    {"year": 2018, "nominal_fare": 2.75, "annual_riders_millions": 3.2, "event": "Residential Towers Overbuild"},
    {"year": 2019, "nominal_fare": 2.75, "annual_riders_millions": 5.9, "event": "Pre-Pandemic Tourism Peak"},
    {"year": 2020, "nominal_fare": 2.75, "annual_riders_millions": 1.5, "event": "COVID-19 Pandemic Shock"},
    {"year": 2021, "nominal_fare": 2.75, "annual_riders_millions": 1.9, "event": "Gradual Hybrid Office Re-entry"},
    {"year": 2022, "nominal_fare": 2.75, "annual_riders_millions": 3.7, "event": "Office Return Mandates"},
    {"year": 2023, "nominal_fare": 2.90, "annual_riders_millions": 5.2, "event": "$2.90 Fare Hike"},
    {"year": 2025, "nominal_fare": 2.90, "annual_riders_millions": 5.5, "event": "Congestion Pricing Implementation"},
    {"year": 2026, "nominal_fare": 3.00, "annual_riders_millions": 6.1, "event": "$3.00 Fare & High-Income Resilience"}
]

df_hy = pd.DataFrame(raw_hudson_yards_data)

# Add local neighborhood economic context (Median Household Income for Chelsea/Hudson Yards)
# This income only applies/displays for years when the station was actually active (2015 onward)
df_hy['local_median_income'] = np.where(df_hy['year'] >= 2015, 150000, np.nan)

print(f"Successfully loaded 34 St-Hudson Yards station dataset spanning {df_hy['year'].min()} to {df_hy['year'].max()}.")
df_hy.tail(11) # Showing the active years with the income data point

## Step 2: Fetch Live Macroeconomic Inflation Data (FRED API)

def fetch_cpi_data():
    """Fetches Annual CPI-U data from FRED API with automatic local fallback."""
    url = "https://api.stlouisfed.org/fred/series/observations"
    params = {
        "series_id": "CPIAUCNS",
        "api_key": "43a00c2d68e0d4d5c96b34b73b5a79a1",
        "file_type": "json"
    }

    try:
        response = requests.get(url, params=params, timeout=10)
        if response.status_code == 200:
            data = response.json()
            obs = data['observations']
            cpi_dict = {}
            for o in obs:
                year = datetime.strptime(o['date'], '%Y-%m-%d').year
                try:
                    val = float(o['value'])
                    if year not in cpi_dict:
                        cpi_dict[year] = []
                    cpi_dict[year].append(val)
                except ValueError:
                    continue
            return {y: np.mean(v) for y, v in cpi_dict.items()}
    except Exception as e:
        print(f"FRED API access encountered an issue: {e}. Executing fallback matrices.")

    # High-precision extended fallback to capture 7-line origins down to active station years
    return {
        1915: 10.1, 1928: 17.1, 1939: 13.9, 1948: 24.1, 1953: 26.7,
        1964: 31.0, 1975: 53.8, 1985: 107.6, 1995: 152.4, 2003: 184.0,
        2015: 237.0, 2016: 240.0, 2017: 245.1, 2018: 251.1, 2019: 255.7,
        2020: 258.8, 2021: 271.0, 2022: 292.7, 2023: 304.7, 2025: 322.1,
        2026: 331.2
    }

cpi_map = fetch_cpi_data()
target_cpi = cpi_map.get(2026, 331.2)

df_hy['cpi'] = df_hy['year'].map(cpi_map)
df_hy['cpi'] = df_hy['cpi'].interpolate(method='linear')

# Calibrate nominal station pricing baseline to 2026 real value terms
df_hy['real_fare_2026_dollars'] = round(df_hy['nominal_fare'] * (target_cpi / df_hy['cpi']), 2)
print("\n--- 34 St-Hudson Yards Real Inflation-Adjusted Price Architecture ---")
print(df_hy[['year', 'nominal_fare', 'real_fare_2026_dollars', 'event']].head())

## Step 3: Compute Price Elasticity of Demand ($E_d$) Timeline

# Using the Midpoint (Arc) Elasticity formula:
# Ed = ((Q2 - Q1) / [(Q1 + Q2) / 2]) / ((P2 - P1) / [(P1 + P2) / 2])

df_hy['pct_chg_ridership'] = df_hy['annual_riders_millions'].pct_change()
df_hy['pct_chg_real_price'] = df_hy['real_fare_2026_dollars'].pct_change()

elasticities = [np.nan] # First entry remains empty

for i in range(1, len(df_hy)):
    q1, q2 = df_hy.loc[i-1, 'annual_riders_millions'], df_hy.loc[i, 'annual_riders_millions']
    p1, p2 = df_hy.loc[i-1, 'real_fare_2026_dollars'], df_hy.loc[i, 'real_fare_2026_dollars']

    # Avoid mathematical errors if prices didn't shift or if there was no ridership (pre-2015)
    if p1 == p2 or (q1 == 0.0 and q2 == 0.0):
        elasticities.append(0.0)
        continue

    delta_q = (q2 - q1) / ((q1 + q2) / 2)
    delta_p = (p2 - p1) / ((p1 + p2) / 2)

    arc_ed = delta_q / delta_p
    elasticities.append(round(arc_ed, 3))

df_hy['price_elasticity'] = elasticities

print("\n--- Calculated 34 St-Hudson Yards Arc Elasticity Metrics ---")
print(df_hy[['year', 'event', 'real_fare_2026_dollars', 'price_elasticity']].to_string())

## Step 4: System Visualization Generation

fig, ax1 = plt.subplots(figsize=(15, 8))

# Plot Real vs Nominal Fare History on primary Y axis
ax1.set_xlabel('Historical Milestones (Year)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Fare Price ($)', color='black', fontsize=12)
ax1.plot(df_hy['year'], df_hy['nominal_fare'], marker='o', label='Nominal Fare', color='#A7A9AC', linestyle='--')
ax1.plot(df_hy['year'], df_hy['real_fare_2026_dollars'], marker='s', label='Real Fare (2026 Inflation Adjusted)', color='#0039A6', linewidth=2)
ax1.tick_params(axis='y', labelcolor='black')
ax1.yaxis.set_major_formatter('${x:,.2f}')

# Plot Hudson Yards Station Ridership Trends on secondary Y axis
ax2 = ax1.twinx()
color_hy = '#B933AD' # 7 Train Purple
ax2.set_ylabel('Annual Station Entries (Millions)', color=color_hy, fontsize=12, fontweight='bold')
ax2.fill_between(df_hy['year'], df_hy['annual_riders_millions'], alpha=0.15, color=color_hy)
ax2.plot(df_hy['year'], df_hy['annual_riders_millions'], marker='d', color=color_hy, linewidth=2.5, label='Hudson Yards Entries (Millions)')
ax2.tick_params(axis='y', labelcolor=color_hy)

# Plot Local Neighborhood Income Context on a tertiary Y axis (displaced to the right)
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60)) # Offset the axis so it doesn't overlap
color_income = '#2CA02C' # Rich Green
ax3.set_ylabel('Local Area Median Income ($)', color=color_income, fontsize=12, fontweight='bold')
# Draw a dashed reference line showcasing the local high-income demographic during active years
ax3.plot(df_hy['year'], df_hy['local_median_income'], color=color_income, linestyle=':', linewidth=3, label='Area MHI (~$150k)')
ax3.tick_params(axis='y', labelcolor=color_income)
ax3.yaxis.set_major_formatter('${x:,.0f}')
ax3.set_ylim(0, 180000)

# Overlay Key Station-Specific Annotations
for i, row in df_hy.iterrows():
    # Only annotate major historic landmarks and modern active years to keep graph legible
    if row['year'] in [1915, 1948, 1975, 1995, 2015, 2019, 2020, 2023, 2026]:
        ax1.annotate(f"{row['year']}: {row['event']}",
                     (row['year'], row['real_fare_2026_dollars']),
                     textcoords="offset points",
                     xytext=(10, 15 if i%2==0 else -25),
                     ha='left', fontsize=8, alpha=0.9,
                     bbox=dict(boxstyle="round,pad=0.3", fc="#FFF3CD", alpha=0.8, ec="#FFEBAA"))

# Compile legends from all three independent axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(lines1 + lines2 + lines3, labels1 + labels2 + labels3, loc='upper left')

plt.title("34 St-Hudson Yards Station: Fare Hikes, Ridership & Local Affluence (1915-2026)", fontsize=15, fontweight='bold', pad=20)
fig.tight_layout()
plt.show()
```
